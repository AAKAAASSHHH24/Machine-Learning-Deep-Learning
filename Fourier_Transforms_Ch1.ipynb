```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pi = np.pi

t_axis = np.linspace(0,4*pi, 100)

f_t = np.ones(len(t_axis))*3

maxs = [(pi,2*pi), (3*pi, 4*pi)]

f_t[(t_axis>=pi) & (t_axis<=2*pi) ] = 0
f_t[(t_axis>=3*pi) & (t_axis<=4*pi)] = 0 
```

```python
plt.plot(t_axis, f_t)
```

```python
def fourier_approx(num_terms = 5, t=t_axis):

  y = 3/2
  
  for n in range(1,num_terms+1):

    if n%2 != 0:

      y_n = (6/n/np.pi)*np.sin(n*t) 

      y += y_n
  
  return y

```

```python
y_f = fourier_approx(5)

plt.plot(t_axis, f_t)
plt.plot(t_axis, y_f)
```

```python
y_f = fourier_approx(10)

plt.plot(t_axis, f_t)
plt.plot(t_axis, y_f)
```

```python
y_f = fourier_approx(15)

plt.plot(t_axis, f_t)
plt.plot(t_axis, y_f)
```

```python
plt.style.use('default')
fig, ax = plt.subplots(2,3,True, True)
fig.set_figheight(8) ; fig.set_figwidth(20)


num_of_terms = [2,3,5]
num_of_terms1 = [10,15,20]

for i in range(3):

  if i<=2:
    ax[0,i].plot(t_axis, f_t, label='Actual Function', color='k', lw=3)
    ax[0,i].plot(t_axis, fourier_approx(num_of_terms[i]), label='Fourier Approximation', color='red')
    ax[0,i].set_title(f'Number of Terms = {num_of_terms[i]}')
    ax[0,i].legend()
    ax[0,i].grid()

for i in range(3): 


    ax[1,i].plot(t_axis, f_t, label='Actual Function', color='k', lw=3)
    ax[1,i].plot(t_axis, fourier_approx(num_of_terms1[i]), label='Fourier Approximation', color='red')
    ax[1,i].set_title(f'Number of Terms = {num_of_terms1[i]}')
    ax[1,i].legend()
    ax[1,i].grid()


plt.suptitle('Approximating a function with Sinusoids : Fourier Transforms', color='darkgreen', size=20)
```
