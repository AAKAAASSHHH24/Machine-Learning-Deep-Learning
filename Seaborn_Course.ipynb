```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## 2. Bi-Variate Analysis : Regression Plots. 

> regplots

>lmplots : same, but more powerful

```python
sns.set_style('darkgrid')
sns.regplot(data=df, x='sepal_width', y='sepal_length')
```

```python
sns.set_style('darkgrid')
sns.lmplot(data=df, x='sepal_width', y='sepal_length')
```

### A. Plots to see Data Points for each category. : Swarmplot & Stripplot

```python
sns.stripplot(data=df, x='sepal_length', y='species')
```

Jitter basically shifts the datapoints up in y-direction keeping x direction constant so that theyre visible. 

```python
sns.stripplot(data=df, x='sepal_length', y='species', jitter=True)
```

#### Swarmplot basically makes sure no two data points sit on top of each other or are showing as a single point. It shows EVERY SINGLE POINT!

```python
sns.swarmplot(data=df, x='sepal_length', y='species')
```

### B. Plot to show distribution of Each category : BoxPlot and Violin plot. 



```python
sns.boxplot(data=df, x='sepal_length', y='species')
```

```python
sns.violinplot(data=df, x='sepal_length', y='species')
```

## boxen : works well with Larger datasets and is a hybrid bw boxplots and violinplots.

```python
sns.boxenplot(data=df, x='sepal_length', y='species')
```

### barplot : statistical estimates of each category

```python
sns.barplot(data=df, x='sepal_length', y='species')
```

```python
def CI95_returner(A):

  min_ = np.mean(A) - 1.96*np.std(A)
  max_ = np.mean(A) + 1.96*np.std(A)

  return min_,max_

mi, ma = CI95_returner(df[df['species']=='versicolor']['sepal_length'])
```

```python
sns.pointplot(data=df, x='sepal_length', y='species',ci=95)
```

```python
sns.displot(df[df['species']=='versicolor']['sepal_length'], kde=True)

plt.axvline(df[df['species']=='versicolor']['sepal_length'].mean(), color='red')
plt.axvline(mi, color='red')
plt.axvline(ma, color='red')
```
