```python
grmax = max(df['GR'])
grmin = min(df['GR'])

# Gr = s*GRmax + (1-s)*GRmin
# s = (GR - GRMin)/(Grmax - Grmin)
```

```python
df['Vsh'] = (df['GR'] - grmin)/(grmax - grmin)
```

```python
plt.style.use('default')
plt.figure(figsize=(4,8))


# plt.plot(df['SP'],df['Depth'],label='SP')
plt.plot(df['GR'],df['Depth'],label='GR',alpha=0.2)
plt.plot(df['Vsh']*100,df['Depth'],label='Vsh')


plt.legend(loc ='upper right')

# plt.axhspan(1500,2500,alpha=0.2,color='gold')

plt.ylim(5000,0)
plt.grid()
```
