CLT states that if we take samples from a distribution with a sample size large enough, the mean of these samples will always follow a Normal Distribution No matter what the original Population Distribution was. 

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
x = np.random.random_sample(2000)
```

```python
sns.distplot(x)

plt.grid()
```

```python
y = []
for i in range(len(x)):
    
    m = np.random.choice(x,50).mean()
    
    y.append(m)
    
y = np.array(y)
```

```python
sns.distplot(y)
plt.grid()
```

```python
fig,ax = plt.subplots(nrows=1,ncols=2)
fig.set_figheight(6)
fig.set_figwidth(15)

sns.distplot(x,ax=ax[0],color='darkorange');
ax[0].grid();
ax[0].set_title('\n\nA Random Distribution-X');

sns.distplot(y,ax=ax[1],color='green');
ax[1].grid();
ax[1].set_title('\n\nDistribution of Sample Means (large sample size)- X(avg)');


plt.suptitle('AN ILLUSTRATION ON THE CENTRAL LIMIT THEOREM WITH PYTHON.');
```
